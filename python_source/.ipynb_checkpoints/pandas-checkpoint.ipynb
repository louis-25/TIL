```python
#append - 행합침
#열합침

df1 = pd.DataFrame({'고객번호':[1001,1002,1003,1004,1005,1006,1007],
                   '이름':['둘리','도우너','또치','길동','희동','마이콜','영희']
                   })
df1
```

```python
df2 = pd.DataFrame({'고객번호':[1001,1001, 1005,1006,1008, 1001],
                   '금액':[10000, 2000, 15000, 5000, 10000, 30000]
                   })
df2
```

```python
#고객번호 존재하는 값을 가졌을 경우에만
df3 = pd.merge(df1, df2)
df3
```

```python
df4 = pd.merge(df1, df2, how='outer') #df1 등장, 또는 df2 등장
df4
#Not a number, NA, null
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
df4 = pd.merge(df1, df2, how="outer")
print(df4)
print(df4.info())
```

```python
print(df4.isnull().sum()) # null의 갯수가 몇개인지 알려줌
df5 = df4['금액'].fillna(0)
print(df5)
df6 = df4['이름'].fillna("==") #null데이터를 '=='로 바꾼다
print(df6)

print(df4['금액'].dropna()) #null데이터가 나오는 행삭제
print(df4.dropna(axis=1)) #null데이터가 나오는 열삭제
```

```python
data = {
    '도시':['서울', '서울', '서울', '부산', '부산', '부산', '인천', '인천'],
    '연도':['2015', '2010','2005','2015','2010','2005','2015','2010'],
    '인구':[99393938, 99393938,  99393938, 99393938, 99393938, 99393938, 99393938, 99393938],
    '지역':['수도권', '수도권', '수도권', '경상권', '경상권', '경상권', '수도권', '수도권']
}
df6 = pd.DataFrame(data)
df6
```

```python
#도시 - 행 / 연도 - 열 / 인구수
df6.pivot('도시', '연도', '인구')
```

```python
df6.pivot_table('인구', '도시', '연도')
```

```python
#aggfunc값이 없으면 기본적으로 평균값으로 나온다
df6.pivot_table('인구', '도시', '연도', margins_name="평균", margins=True)
```

```python
#합계
df6.pivot_table('인구', '도시', '연도', aggfunc=sum, margins_name="합계", margins=True)
```
